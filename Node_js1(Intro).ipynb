# **✅ What is Node.js?**
Node.js is a tool that lets you run JavaScript on your computer or server, not just inside a browser. It’s great for building backend/server-side applications.

**JavaScript Runtime:**

Node.js is an open-source, cross-platform environment.

It allows you to run JavaScript code outside of the browser—for example, on a server.

**Different Environment:**

Node.js enables running JavaScript in a non-browser environment, such as on a server or a computer.

**Built on Chrome's V8 Engine:**

Node.js uses the V8 JavaScript engine (used by Google Chrome).

V8 compiles JavaScript directly into native machine code, making it very fast.

**V8 is written in C++:**

The V8 engine is developed in C++ for high performance.

**V8 + Backend Features = NodeJS:**

Node.js combines the V8 engine with backend capabilities (like file system access, network tools, etc.), forming a powerful runtime environment.

# **Design**
Node.js is event-driven and uses a non-blocking system.
➤ This means it can handle many tasks at the same time, making it fast and efficient.

# **Full-Stack JavaScript**
You can use JavaScript on both frontend (client) and backend (server).
➤ So, you only need to learn one language for the whole app!

# **Scalability**
Node.js is great for building apps that need to handle a lot of users or data.
➤ It’s built to scale easily.

# **Versatility**
Node.js works well for different kinds of apps like:
➤ Websites, real-time chats, and REST APIs.



# **🟢 Node.js Features**

**Non-blocking I/O**

Node.js can do many tasks at the same time without waiting.

This makes it fast and efficient, especially for apps that read/write a lot (like file or database access).

**Networking Support**

Node.js can handle TCP/UDP connections.

This means it can build low-level network apps, like chat apps or custom servers—things normal browsers can’t do.

# **🟢 More Features of Node.js**

# **File System Access**

Node.js can read and write files on your computer.

This is not allowed in browsers for security, but Node.js gives you this power.

# **Server-Side Capabilities**

Node.js lets JavaScript run on the server, not just in the browser.

You can:

Handle HTTP requests

Do file operations

Run backend logic (like connecting to a database)

# **Modules**

You can split your code into smaller, reusable parts (called modules).

Use require() to bring those modules into your app.

Makes your code organized and easy to manage.

# **❌ Node.js Features That Are Not Available (Compared to Browsers)**

**Window Object**

The global window object (used in browsers) does not exist in Node.js.

Node runs outside the browser, so you don’t need window.

**DOM Manipulation**

You can't use document.getElementById() or change HTML.

Node.js has no built-in DOM, because it's not meant for webpage interaction.

**BOM (Browser Object Model)**

Objects like navigator, location, screen, etc., are only in browsers, not in Node.js.

**Web-Specific APIs**

These APIs are not available in Node.js:

localStorage

sessionStorage

cookies (in the browser sense)

fetch (browser-based version)

Node uses its own packages for HTTP requests (like axios or http module) and cookie handling.

# **🖥️ JavaScript in Browser (Client-Side)**
Used for:

✅ Display & render web pages

✅ Handle user interaction (clicks, inputs)

✅ DOM manipulation (e.g., changing HTML content)

✅ Fetch data from server

✅ Animations, UI behavior

⛔ Cannot:

Access file systems

Connect directly to databases

Manage sessions on its own

# **🔧 JavaScript in Node.js (Server-Side)**

Used for:

✅ Database management
(CRUD operations with MongoDB, MySQL, etc.)

✅ Authentication & authorization
(e.g., verifying login info, checking access rights)

✅ Input validation
(Sanitizing form data before saving to DB)

✅ Session management
(Tracking user login status with cookies or tokens)

✅ Business logic & APIs
(Respond to requests, process data, etc.)

⛔ Cannot:

Access the DOM (document, window)

Directly interact with browser elements

Use localStorage / sessionStorage

**🔁 API Management**

Provides and handles interfaces for applications to interact, ensuring smooth data exchange and integration.

**🚫 Error Handling**

Manages and responds to errors effectively to maintain system stability and provide useful error messages.

**🛡️ Security Measures**

Implements protocols to protect data from unauthorized access and attacks, such as SQL injection and cross-site scripting (XSS).

**🔐 Data Encryption**

Secures sensitive information by encrypting data stored in databases and during transmission.

**📊 Logging and Monitoring**

Keeps records of system activity to diagnose issues and monitor system health and security.
