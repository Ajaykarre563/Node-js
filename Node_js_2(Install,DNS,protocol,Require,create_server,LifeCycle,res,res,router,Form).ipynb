# **What is mean by DNS?**

image.png

# **Domain Name Entry**

The user types a domain name (e.g., www.example.com) into the browser.

**DNS Query**

The browser sends a DNS query to get the corresponding IP address for that domain.

**DNS Server**

The DNS server finds and returns the correct IP address (e.g., 87.245.200.153) for the domain.

**Browser Connects**

Using the IP address, the browser connects to the web server and loads the website.



image.png

# **Root DNS**

Acts as the starting point for DNS resolution.

It doesn’t know the final IP, but it redirects the query to the appropriate TLD (Top-Level Domain) server.

Example TLDs: .com, .org

# **TLD (Top-Level Domain) DNS**

Handles queries for specific domain extensions like .com, .net, .org.

It then forwards the query to the correct Authoritative DNS server.

# **Authoritative DNS**

This server knows the actual IP address of the domain.

It responds to the DNS query with the IP.

Examples: Cloudflare DNS, Google DNS

# **🔐 1. HTTP (HyperText Transfer Protocol)**
📌 What it does: Sends webpage data in plain text (no encryption).

🛍️ Myntra example: If you visit an old or test version like http://myntra.com, your login or search data is not safe — anyone on the network can read it.

⚠️ Not used today for secure shopping.

# **🔐 2. HTTPS (Secure HTTP)**
📌 What it does: Encrypts your data using SSL/TLS for security.

🛍️ Myntra example: You visit https://www.myntra.com, log in, browse products, and make payments. Your personal info, password, and card details are secure.

✅ Used in all secure websites like Myntra.

# **🔁 3. TCP (Transmission Control Protocol)**
📌 What it does: Ensures data is sent reliably and in order over the internet.

🛍️ Myntra example: When you open the site or place an order, TCP ensures that every image, product detail, and payment request reaches your browser correctly and fully — no missing pieces.

🔄 Works behind the scenes every time you shop online.

# **create server for both req and res**



```
const http = require('http'); // Import the http module

const server = http.createServer((req, res) => {
    console.log(req.url); // Log the requested URL for debugging
    res.writeHead(200, { 'Content-Type': 'text/plain' }); // Set response header
    res.write('Hello, World!'); // Send a response body
    res.end(); // End the response
});

server.listen(3001, () => {
    console.log('Server running on port 3001'); // Confirm server is running
});
```

# **router**



```
const http = require('http');

const server = http.createServer((req, res) => {
    console.log(req.url, req.method, req.headers);

    if (req.url === '/') {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body><h1>Welcome to Home</h1></body>');
        res.write('</html>');
        res.end();
    } else if (req.url === '/products') {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body><h1>Checkout our products!</h1></body>');
        res.write('</html>');
        res.end();
    } else {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body><h1>404 - Page Not Found</h1></body>');
        res.write('</html>');
        res.end();
    }
});

const PORT = 3001;

server.listen(PORT, () => {
    console.log(`Server running on address http://localhost:${PORT}`);
});
```

# **Form code:**



```
const http = require('http');
const querystring = require('querystring');

const server = http.createServer((req, res) => {
    console.log(req.url, req.method, req.headers);

    if (req.url === '/' && req.method === 'GET') {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body>');
        res.write('<h1>Welcome to Home page</h1>');
        res.write('<form action="/submit-details" method="POST">');
        res.write('<input type="text" id="name" name="name" placeholder="Enter your name"><br><br>');
        res.write('<label for="gender">Gender:</label>');
        res.write('<input type="radio" id="male" name="gender" value="male">');
        res.write('<label for="male">Male</label>');
        res.write('<input type="radio" id="female" name="gender" value="female">');
        res.write('<label for="female">Female</label><br><br>');
        res.write('<button type="submit">Submit</button>');
        res.write('</form>');
        res.write('</body>');
        res.write('</html>');
        res.end();
    } else if (req.url === '/submit-details' && req.method === 'POST') {
        let body = '';
        req.on('data', chunk => {
            body += chunk.toString();
        });
        req.on('end', () => {
            const parsedData = querystring.parse(body);
            const name = parsedData.name || 'Unknown';
            const gender = parsedData.gender || 'Not specified';

            res.setHeader('Content-Type', 'text/html');
            res.write('<html>');
            res.write('<head><title>Complete Coding</title></head>');
            res.write('<body>');
            res.write('<h1>Submitted Details</h1>');
            res.write(`<p>Name: ${name}</p>`);
            res.write(`<p>Gender: ${gender}</p>`);
            res.write('</body>');
            res.write('</html>');
            res.end();
        });
    } else if (req.url === '/products') {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body><h1>Checkout our products!</h1></body>');
        res.write('</html>');
        res.end();
    } else {
        res.setHeader('Content-Type', 'text/html');
        res.write('<html>');
        res.write('<head><title>Complete Coding</title></head>');
        res.write('<body><h1>404 - Page Not Found</h1></body>');
        res.write('</html>');
        res.end();
    }
});

const PORT = 3001;

server.listen(PORT, () => {
    console.log(`Server running on address http://localhost:${PORT}`);
});
```
